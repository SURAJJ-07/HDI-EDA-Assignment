```python
hdi_2022_df = hdi_2022_df.drop_duplicates()
```

```python
numeric_cols = ['hdi', 'gross_inc_percap', 'life_expectancy']
hdi_2022_df[numeric_cols] = hdi_2022_df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
hdi_2022_df = hdi_2022_df.dropna(subset=['hdi'])
```

```python
high_hdi = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]
high_hdi = high_hdi.sort_values(by='gross_inc_percap', ascending=False)
high_hdi.head(10)
```

```python
def hdi_category(hdi):
    if hdi < 0.550:
        return "Low"
    elif hdi < 0.700:
        return "Medium"
    elif hdi < 0.800:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(hdi_category)
```

```python
hdi_2022_df['HDI Category'].value_counts()
```

```python
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1b = pd.read_csv("HDI_problem1B.csv")
df_1b.head()
```

```python
south_asia = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

middle_east = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]
```

```python
sa_2020_22 = df_1b[df_1b['country'].isin(south_asia)]
me_2020_22 = df_1b[df_1b['country'].isin(middle_east)]
```

```python
sa_2020_22.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
me_2020_22.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)
```

```python
sa_stats = sa_2020_22['hdi'].agg(['mean', 'std'])
me_stats = me_2020_22['hdi'].agg(['mean', 'std'])

sa_stats, me_stats
```

```python
sa_top = sa_2020_22.sort_values('hdi', ascending=False).head(3)
sa_bottom = sa_2020_22.sort_values('hdi').head(3)

me_top = me_2020_22.sort_values('hdi', ascending=False).head(3)
me_bottom = me_2020_22.sort_values('hdi').head(3)
```

```python
compare_df = pd.concat([sa_top, sa_bottom, me_top, me_bottom])
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=compare_df, x='country', y='hdi', hue='year')
plt.title("Top & Bottom HDI Performers (2020–2022)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.barplot(data=df_1b[df_1b['country'].isin(south_asia + middle_east)],
            x='year', y='gender_development', hue='country')
plt.title("Gender Development Comparison")
plt.show()
```

```python
sns.barplot(data=df_1b[df_1b['country'].isin(south_asia + middle_east)],
            x='year', y='life_expectancy', hue='country')
plt.title("Life Expectancy Comparison")
plt.show()
```

```python
sns.barplot(data=df_1b[df_1b['country'].isin(south_asia + middle_east)],
            x='year', y='gross_inc_percap', hue='country')
plt.title("GNI per Capita Comparison")
plt.show()
```

```python
def disparity_stats(df):
    mean = df['hdi'].mean()
    std = df['hdi'].std()
    hdi_range = df['hdi'].max() - df['hdi'].min()
    cv = std / mean
    return mean, std, hdi_range, cv

sa_disp = disparity_stats(sa_2020_22)
me_disp = disparity_stats(me_2020_22)

sa_disp, me_disp
```

```python
sa_corr = sa_2020_22[['hdi', 'life_expectancy']].corr()
me_corr = me_2020_22[['hdi', 'life_expectancy']].corr()

sa_corr, me_corr
```

```python
sns.lmplot(data=sa_2020_22, x='life_expectancy', y='hdi')
plt.title("South Asia: HDI vs Life Expectancy")
plt.show()
```

```python
sns.lmplot(data=me_2020_22, x='life_expectancy', y='hdi')
plt.title("Middle East: HDI vs Life Expectancy")
plt.show()
```

```python
Q1 = df_1b['hdi'].quantile(0.25)
Q3 = df_1b['hdi'].quantile(0.75)
IQR = Q3 - Q1

outliers = df_1b[(df_1b['hdi'] < Q1 - 1.5*IQR) | (df_1b['hdi'] > Q3 + 1.5*IQR)]
outliers
```

```python
sns.scatterplot(data=df_1b, x='gross_inc_percap', y='hdi')
sns.scatterplot(data=outliers, x='gross_inc_percap', y='hdi', color='red')
plt.title("Outliers in HDI vs GNI")
plt.show()
```
